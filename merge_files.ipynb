## Import

```python
import pandas as pd
```

## Load files

```python
content_df = pd.read_csv('Content_preprocessed.csv')
reactions_df = pd.read_csv('Reaction_preprocessed.csv')
reactionType_df = pd.read_csv('ReactionTypes.csv')
```

```python
reactions_df.head()
```

### Merge Content onto Reactions

```python
for i in range(len(reactions_df)):
    reactions_df.loc[i, 'Content Type'] = content_df['Type'][content_df['Content ID'] == reactions_df.loc[i, 'Content ID']].values[0]
    reactions_df.loc[i, ['Category']] = content_df['Category'][content_df['Content ID'] == reactions_df.loc[i, 'Content ID']].values[0]
    #print(content_df['Type'][content_df['Content ID'] == reactions_df.loc[i, 'Content ID']].values[0])
```

```python
reactions_df.head(10)
```

### Merge Reaction Types onto Reactions

```python
for i in range(len(reactions_df)):
    reactions_df.loc[i, ['Sentiment']] = reactionType_df['Sentiment'][reactionType_df['Reaction Type'].values == reactions_df.loc[i, ['Reaction Type']].values].values[0]
    reactions_df.loc[i, ['Score']] = reactionType_df['Score'][reactionType_df['Reaction Type'].values == reactions_df.loc[i, ['Reaction Type']].values].values[0]
```
